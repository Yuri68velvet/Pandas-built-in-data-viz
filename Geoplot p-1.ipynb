```python
import plotly.plotly as py
```

```python
import plotly.graph_objs as go
```

```python
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
```

```python
init_notebook_mode(connected=True)
```

```python
import pandas as pd
```

```python
df=pd.read_csv('countries-of-the-world.zip')
df.head()
```

```python
ls
```

```python
data=dict(
    type='choropleth',
    locations=df['Climate'],
    z=df['Population'],
    colorscale='Blackbody',
    text=df['Population'],
    colorbar={'title':'GDP of the world'}
    
    )
```

```python
layout=dict(
    title='GDP of the world',
    geo=dict(
    showframe=False,
    projection={'type':'mercator'})
    )
```

```python
choromap=go.Figure(data=[data],layout=layout)
iplot(choromap)
```

```python
data1=pd.read_csv('world-happiness-report-2019.zip')
data1.head()
```

```python
data2=dict(type='choropleth',
         locations=data1['Country (region)'],
          colorscale='Blackbody',
          text=data1['Ladder'],
          z=data1['Social support'],
         colorbar={'title':'Social support of the world'})
```
