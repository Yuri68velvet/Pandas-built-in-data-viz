# Fake data

```python
df=pd.DataFrame(np.random.randn(100,4),columns='A B C D'.split())
```

```python
df.head()
```

```python
df2=pd.DataFrame({'Category':['A','B','C',],
                  'Values':[32,42,50]})
```

```python
df2
```

# Using Cufflinks and plot

* scatter
* bar
* box
* spread
* ratio
* heatmap
* surface
* histograms
* bubble

# Scatter

```python
df.plot(kind='scatter',x='A',y='B',s=10)
```

```python
df.iplot(kind='scatter',x='A',y='B',mode='markers',size=5)
```

# Bar Plots

```python
df2.plot(kind='bar',x='Category',y='Values')
```

```python
df2.iplot(kind='bar',x='Category',y='Values')
```

```python
df.count().iplot(kind='bar')
```

# Boxplot

```python
df.iplot(kind='box')
```

# 3D surface

```python
df3=pd.DataFrame({'x':[1,2,3,4,5],'y':[1,2,3,4,5],'z':[0,0,0,0,0]})
df3.iplot(kind='surface')
```

```python

df3.iplot(kind='surface',colorscale='rdylbu')
```

# Spread

```python
df[['A','B']].iplot(kind='spread')
```
