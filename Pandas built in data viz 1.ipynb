# Plot types

#### there are several plot types built-in pandas,　most of them statistical plots by nature:


* df.plot.area
* df.plot.bar
* df.plot.density
* df.plot.hist
* df.plot.line
* df.plot.scatter
* df.plot.bar
* df.plot.box
* df.plot.kde
* df.plot.pie

You can also just call df.plot(kind=hist) pr replace that kind argument with any of the key terms shown inthe list above(e.g.'box','bath etc)')

# Scatter Plots

```python
df1.plot.scatter(x='A',y='B')
```

```python
df2.plot(kind='scatter',x='a',y='b')
```

#### You can use c to color based off another column value Use cmap to indicate colormap to use.
For all color maps,check out:
    https://matplotlib.org/3.1.0/tutorials/colors/colormaps.html

```python
df1.plot.scatter(x='A',y='B',c='C',cmap='viridis')
```

#### Or use s to indicate size based off another column.s parameter needs to be an array,not just name of a column:

```python
df1.plot.scatter(x='A',y='B',s=df1['C']*100)
```

# Box plot

```python
df2.plot.box()
```

# Hexagonal Bin plot
Useful for Bivariate Data,alternative to scatterplot:

```python
df=pd.DataFrame(np.random.randn(1000,2),columns=['a','b'])
df.plot.hexbin(x='a',y='b',gridsize=25)
```
